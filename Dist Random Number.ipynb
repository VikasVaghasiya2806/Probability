```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random integers between 10 and 100 (inclusive)
random_numbers = np.random.randint(10, 2000, size=1000)

# Plot the distribution
plt.hist(random_numbers, bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Random Integers (10 to 100)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

# Seed for reproducibility
np.random.seed(42)

# Generate first dataset: integers between 10 and 100
random_numbers1 = np.random.randint(10, 101, size=1000)

# Generate second dataset: integers between 50 and 150
random_numbers2 = np.random.randint(10, 101, size=1000)

# Compute statistics for dataset 1
mean1 = np.mean(random_numbers1)
median1 = np.median(random_numbers1)
std1 = np.std(random_numbers1)
min1 = np.min(random_numbers1)
max1 = np.max(random_numbers1)

print("Dataset 1 Statistics:")
print(f"Mean: {mean1:.2f}")
print(f"Median: {median1:.2f}")
print(f"Standard Deviation: {std1:.2f}")
print(f"Min: {min1}")
print(f"Max: {max1}")

# Compute statistics for dataset 2
mean2 = np.mean(random_numbers2)
median2 = np.median(random_numbers2)
std2 = np.std(random_numbers2)
min2 = np.min(random_numbers2)
max2 = np.max(random_numbers2)

print("\nDataset 2 Statistics:")
print(f"Mean: {mean2:.2f}")
print(f"Median: {median2:.2f}")
print(f"Standard Deviation: {std2:.2f}")
print(f"Min: {min2}")
print(f"Max: {max2}")

# Plot histogram with KDE for dataset 1
plt.figure(figsize=(10, 5))
sns.histplot(random_numbers1, bins=20, kde=True, color='skyblue', edgecolor='black')
plt.title('Histogram with Density Curve (Dataset 1)')
plt.xlabel('Random Value')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()


# Compare histograms of both datasets
plt.figure(figsize=(10, 5))
plt.hist(random_numbers1, bins=20, alpha=0.5, label='10-100')
plt.hist(random_numbers2, bins=20, alpha=0.5, label='10-100')
plt.title('Comparison of Random Integer Distributions')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random pixel values for a 100x100 RGB image
random_image = np.random.randint(0, 256, size=(100, 100, 3), dtype=np.uint8)

# Display the image
plt.imshow(random_image)
plt.title('Random RGB Image')
plt.axis('off')
plt.show()
```

```python

# Parameters for Gaussian distribution
mean = 128      # center brightness
std_dev = 500    # spread of brightness

# Generate Gaussian-distributed pixel values (grayscale)
gaussian_image = np.random.normal(loc=mean, scale=std_dev, size=(100, 100))

# Clip values to the valid range 0-255
gaussian_image = np.clip(gaussian_image, 0, 255)

# Convert to unsigned 8-bit integer
gaussian_image = gaussian_image.astype(np.uint8)

# Display the grayscale image
plt.imshow(gaussian_image, cmap='gray')
plt.title('Random Gaussian Noise Image')
plt.axis('off')
plt.show()
```

```python
mean = 128      # center brightness
std_dev = 50    # spread of brightness

# Generate Gaussian-distributed pixel values for RGB
gaussian_rgb = np.random.normal(loc=mean, scale=std_dev, size=(100, 100, 3))

# Clip to 0255
gaussian_rgb = np.clip(gaussian_rgb, 0, 255)

# Convert to uint8
gaussian_rgb = gaussian_rgb.astype(np.uint8)

# Display the RGB image
plt.imshow(gaussian_rgb)
plt.title('Random Gaussian RGB Image')
plt.axis('off')
plt.show()
```
