```python
import matplotlib.pyplot as plt

# Discrete outcomes
outcomes = [1, 2, 3, 4, 5, 6]
probabilities = [1/6] * 6  # equal probability for fair die

# Plotting PDF
plt.figure(figsize=(8, 4))
plt.bar(outcomes, probabilities, color='skyblue', edgecolor='black')
plt.title("PDF - Discrete Uniform Distribution (Fair Die)")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.grid(axis='y')
plt.show()
```

```python
# Manual CDF calculation
cumulative_probs = []
total = 0
for p in probabilities:
    total += p
    cumulative_probs.append(total)

# Plotting CDF
plt.figure(figsize=(8, 4))
plt.step(outcomes, cumulative_probs, where='post', color='orange')
plt.title("CDF - Cumulative Distribution Function")
plt.xlabel("Outcome")
plt.ylabel("Cumulative Probability")
plt.grid()
plt.show()
```

```python
# Manual Expectation
expectation = 0
for i in range(len(outcomes)):
    expectation += outcomes[i] * probabilities[i]

# Manual E[X^2]
ex2 = 0
for i in range(len(outcomes)):
    ex2 += (outcomes[i] ** 2) * probabilities[i]

# Variance
variance = ex2 - (expectation ** 2)

# Standard Deviation
std_dev = variance ** 0.5

print("📌 Expectation E[X] =", expectation)
print("📌 Variance =", variance)
print("📌 Standard Deviation =", std_dev)
```

```python
# A biased die where 6 is more likely
outcomes = [1, 2, 3, 4, 5, 6]
probabilities = [0.05, 0.1, 0.1, 0.2, 0.25, 0.3]

# Recalculate Expectation and Variance
expectation = sum(outcomes[i] * probabilities[i] for i in range(len(outcomes)))
ex2 = sum((outcomes[i] ** 2) * probabilities[i] for i in range(len(outcomes)))
variance = ex2 - (expectation ** 2)
std_dev = variance ** 0.5

print("➡️ Custom PDF Expectation:", expectation)
print("➡️ Custom PDF Variance:", variance)
print("➡️ Custom PDF Std Dev:", std_dev)

# Plot the biased PDF
plt.figure(figsize=(8, 4))
plt.bar(outcomes, probabilities, color='lightcoral', edgecolor='black')
plt.title("PDF - Biased Die Distribution")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.grid(axis='y')
plt.show()
```

```python
# CDF for biased die
cumulative_probs = []
total = 0
for p in probabilities:
    total += p
    cumulative_probs.append(total)

# Plot both
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.bar(outcomes, probabilities, color='mediumseagreen')
plt.title("PDF (Biased Die)")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.grid(axis='y')

plt.subplot(1, 2, 2)
plt.step(outcomes, cumulative_probs, where='post', color='darkorange')
plt.title("CDF (Biased Die)")
plt.xlabel("Outcome")
plt.ylabel("Cumulative Probability")
plt.grid()

plt.tight_layout()
plt.show()
```
